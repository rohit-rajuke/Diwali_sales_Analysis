## Data Analysis

## Gender

```python
aa = sb.countplot(x='Gender',data = df)
for bars in aa.containers:
    aa.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sb.barplot(x = 'Gender',y = 'Amount',data = sales_gen)
```

#### Most of the buyers are female as compare to men 

## Age

```python
bb = sb.countplot(data = df,x = 'Age Group',hue = 'Gender')
for bars in bb.containers:
    bb.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sb.barplot(x='Age Group',y='Amount',data = sales_age)
```

#### Most of the buyers are of age group between 26-35 yrs female

## State

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sb.set(rc={'figure.figsize':(15,5)})
sb.barplot(data = sales_state, x = 'State',y='Orders')
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sb.set(rc={'figure.figsize':(15,5)})
sb.barplot(data = sales_state, x = 'State',y='Amount')
```

#### Most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

## Marital Status

```python
marital_status_counts = df['Marital_Status'].value_counts()
colors = ['orange', 'Cyan','orange','Cyan']

plt.figure(figsize=(7, 5))
plt.pie(marital_status_counts, labels=marital_status_counts.index, autopct='%1.1f%%', startangle=140,colors=colors)
plt.axis('equal') 
plt.title('Distribution of Marital Status')
plt.show()
```

```python
sales_state = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sb.set(rc={'figure.figsize':(6,5)})
sb.barplot(data = sales_state, x='Marital_Status',y='Amount',hue='Gender')
```

#### Most of the buyers are married (women) and they have high purchasing.

## Occupation

```python
sb.set(rc={'figure.figsize':(20,5)})
dd = sb.countplot(data=df,x = 'Occupation')
for bars in dd.containers:
    dd.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sb.set(rc={'figure.figsize':(20,5)})
sb.barplot(data = sales_state, x='Occupation',y='Amount')
```

#### Most of the buyers are working in IT, Healthcare and Aviation sector.

## Product Category

```python
sb.set(rc={'figure.figsize':(20,5)})
dd = sb.countplot(data=df,x = 'Product_Category')
for bars in dd.containers:
    dd.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sb.set(rc={'figure.figsize':(20,5)})
sb.barplot(data = sales_state, x='Product_Category',y='Amount')
```

#### Most of the sold products are from Food, Clothing and Electronics category.

```python
sales_state = df.groupby(['Product_ID','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sb.set(rc={'figure.figsize':(20,5)})
sb.barplot(data = sales_state, x='Product_ID',y='Amount')
```
